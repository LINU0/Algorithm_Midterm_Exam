```python
import time
import matplotlib.pyplot as plt
```

```python
n_list=list(range(10, 101, 10))
```

```python
def fibonacci(n):
  if n <= 1:
    return n
  else:
    return(fibonacci(n-1)+fibonacci(n-2))
```

```python
def fibonacci_dp(n):
  if n <= 1:
    return n
  else:
    table=[0,1]
    for i in range(2,n+1):
      table.append(table[i-1]+table[i-2])
    return table[n]
```

```python
#執行3小時強制終止
execution_time_list=[]    
for n in n_list:
  print(n)
  start_time = time.process_time()
  fibonacci(n)
  end_time = time.process_time()
  execution_time = end_time - start_time
  execution_time_list.append(execution_time)
```

```python
execution_time_list
'''執行結果[4.8207999999050344e-05,
 0.0036538770000014154,
 0.5439417609999992,
 39.522097556000006,
 4731.131523279]'''
```

```python
dp_execution_time_list=[]
for n in n_list:
  dp_start_time = time.process_time()
  print(n,fibonacci_dp(n))
  dp_end_time = time.process_time()
  dp_execution_time = dp_end_time - dp_start_time
  dp_execution_time_list.append(dp_execution_time)
```

```python
plt.plot(n_list[:5], execution_time_list, label="fibonacci")
plt.plot(n_list[:5], dp_execution_time_list[:5], label="fibonacci_dp")

plt.xlabel("n")
plt.ylabel("Execution Time")
plt.title("Execution Time for F(n)")
plt.legend()
plt.show()
```
